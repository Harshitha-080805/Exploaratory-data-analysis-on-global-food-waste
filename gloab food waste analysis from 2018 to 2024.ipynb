## Visualization 

### 1. Bar Plot â€“ Total Waste by Country

```python
waste_by_country = data.groupby('Country')['Total Waste (Tons)'].sum()
waste_by_country.plot(kind='bar', figsize=(10, 5), color='skyblue')
plt.title('Total Food Waste by Country')
plt.ylabel('Waste (Tons)')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 2. Line Plot â€“ Economic Loss over Years

```python
loss_by_year = data.groupby('Year')['Economic Loss (Million $)'].sum()
loss_by_year.plot(kind='line', marker='o', figsize=(8, 4), color='green')
plt.title('Economic Loss Over Years')
plt.ylabel('Economic Loss (Million $)')
plt.xlabel('Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 3. Pie Chart â€“ Food Waste by Category

```python
waste_by_category = data.groupby('Food Category')['Total Waste (Tons)'].sum()

waste_by_category.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6, 6))
plt.title('Food Waste Share by Category')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

### 4. Histogram â€“ Distribution of Avg Waste per Capita

```python
data['Avg Waste per Capita (Kg)'].plot(kind='hist', bins=10, color='orange', edgecolor='black')
plt.title('Distribution of Avg Waste per Capita')
plt.xlabel('Kg per Capita')
plt.tight_layout()
plt.show()
```

### 5. Scatter Plot â€“ Economic Loss vs Total Waste

```python
plt.scatter(data['Total Waste (Tons)'], data['Economic Loss (Million $)'], color='purple')
plt.title('Economic Loss vs Total Waste')
plt.xlabel('Total Waste (Tons)')
plt.ylabel('Economic Loss (Million $)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
